```python
#importing lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm #support vector machine
from sklearn.metrics import accuracy_score
```

```python
#load the data of diabates
df=pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
# number of rows and col
df.shape
```

```python
#info of data
df.info()
```

```python
# data stats measure
df.describe()
```

```python
#total values of each outcome
df['Outcome'].value_counts()
```

0 = non-diabetics

1= diabetics

```python
#clculate the mean of each value in each col
df.groupby('Outcome').mean()
```

```python
#separating the data and labels
X=df.drop(columns='Outcome',axis=1)
Y=df['Outcome']
```

```python
#standardized the data
#standardscaler help differnet type of data to compare by making measurement range is same range in each cloumn
# for eg wight is measure in kg and height in feet. so range of weight vary in high side than height. so, to compare them they need to in same range
scalar=StandardScaler() # use when there is many varation of data to be calculated
scalar.fit(X)
standardized_data= scalar.transform(X)
#can also use .fit_transform(x)
```

```python
#train and test data split
X_train,X_test,Y_train,Y_test=train_test_split(standardized_data,Y,test_size=0.2,stratify=Y,random_state=2) #stratify make data to be seperated in a way where that in include the value of each group euall in both train and test data
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

```python
#training the model
#using kernel rbf increase the dimesion
classifier=svm.SVC(kernel='linear')
#classifier=svm.SVC(kernel='rbf') #support vector classifier
```

```python
#training the support vectorr machine classifer
classifier.fit(X_train,Y_train)
```

```python
#model Evaluation
#accuracy score
X_train_prediction=classifier.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction=classifier.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print(test_data_accuracy)
```
